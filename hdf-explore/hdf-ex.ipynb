```python
import h5py
import pandas as pd

filename = "data/1bzc.hdf"

with h5py.File(filename, "r") as pdb:
    # Print all root level object names (aka keys) 
    # these can be group or dataset names 
    print(f"Keys, {len(pdb.keys())}: {pdb.keys()}")
    series_data = pd.Series(pdb.keys(), name="complexes")
    series_data.to_csv("ingenni_complexes.csv", index=False)
    print("-------------------------")
    
    # get first object name/key; may or may NOT be a group
    # key = list(pdb.keys())[0]
    key = "1bzc"

    print(f"Key Type: {type(pdb[key])}") 
    print("-------------------------")

    # If a_group_key is a group name, 
    # this gets the object names in the group and returns as a list
    data1 = list(pdb['1bzc'])


    # If a_group_key is a dataset name, 
    # this gets the dataset values and returns as a list
    # data = list(f[a_group_key])
    print("d1", type(data1), len(data1), data1[:5])


    # If a_group_key is a group name, 
    # this gets the object names in the group and returns as a list
    data2 = list(pdb['1bzc'][data1[0]])


    # If a_group_key is a dataset name, 
    # this gets the dataset values and returns as a list
    # data = list(f[a_group_key])
    print("d2", type(data2), len(data2), data2[:5])

    # If a_group_key is a group name, 
    # this gets the object names in the group and returns as a list
    data3 = list(pdb['1bzc'][data1[0]][data2[0]])


    # If a_group_key is a dataset name, 
    # this gets the dataset values and returns as a list
    # data = list(f[a_group_key])
    print("d3", type(data3), len(data3), data3[:5])

    # If a_group_key is a group name, 
    # this gets the object names in the group and returns as a list
    data4 = list(pdb['1bzc'][data1[0]][data2[0]][data3[0]])


    # If a_group_key is a dataset name, 
    # this gets the dataset values and returns as a list
    # data = list(f[a_group_key])
    print("d4", type(data4), len(data4), data4[:5])

    # If a_group_key is a group name, 
    # this gets the object names in the group and returns as a list
    data5 = list(pdb['1bzc'][data1[0]][data2[0]][data3[0]][data4[0]])


    # If a_group_key is a dataset name, 
    # this gets the dataset values and returns as a list
    # data = list(f[a_group_key])
    print("d5", type(data5), len(data5))

    import numpy as np
    d5_np = np.array(data5)
    print("d5_np", type(d5_np), d5_np.shape)





    # # preferred methods to get dataset values:
    # ds_obj = f[a_group_key]      # returns as a h5py dataset object
    # ds_arr = f[a_group_key][()]  # returns as a numpy array
```
